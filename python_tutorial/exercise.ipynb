### Read CSV file

```python
# read online csv file by using read_csv funtion in pandas library

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data", header=None)
print(df)
```

```python
df.head(5)
```

### Add Header

There is no header for colums attribute, So let's add header to df variable.

```python
# create header lists

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns = headers
df.head(5)
```

Also, we replace "?" to "Nan"

```python
df = df.replace("?", np.nan)
df.head(5)
```
